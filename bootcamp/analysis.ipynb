```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_num is your DataFrame
heatmap_data = df_num.corr()

# Set the size of the heatmap
fig, ax = plt.subplots(figsize=(16, 8))

# Plot the heatmap
plot = sns.heatmap(heatmap_data, annot=True, ax=ax)

# Show the plot
plot.set_title("Facebook Metrics")
plt.show()
```

```python
# Extract date and time
import pandas as pd
from datetime import datetime

# Define a function to extract the date and time from a given string
def extract_date_time(date_string):
    date_object = datetime.strptime(date_string, "%m/%d/%Y %I:%M:%S %p %z")
    return date_object.date(), date_object.time()

# Apply the function to the entire 'Start Date/Time' column
df[['Date', 'Time']] = df['Start Date/Time'].apply(extract_date_time).apply(lambda x: pd.Series(x, index=['Date', 'Time']))

# Display the modified DataFrame
df.head(3)
```

```python
df.info()
```

```python
# Convert 'Date' column to datetime format to extract the month from the date
from datetime import datetime, timedelta
df['revised_date'] = pd.to_datetime(df['Date'])

# Replace the original 'date' column with the 'revised_date' column
df['Date'] = df['revised_date']

df = df.drop(columns=['revised_date'])
```

```python
df.head(5)
```

```python
df.info()
```

```python
# Extract month and year from date column
import calendar

df['Month'] = df['Date'].dt.month.apply(lambda x: calendar.month_name[x])
df['Year'] = df['Date'].dt.year
```

```python
df = df.drop(df[df['Year'] == 2019].index)
```
