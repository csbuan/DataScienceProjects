Conversion of Data Types

```python
# Convert 'Date' column to datetime format to extract the month from the date
from datetime import datetime, timedelta
df['revised_date'] = pd.to_datetime(df['Date'])

# Replace the original 'date' column with the 'revised_date' column
df['Date'] = df['revised_date']

df = df.drop(columns=['revised_date'])
```

```python
# List of columns to convert
columns_to_convert = [' Manufacturing Price ', ' Sale Price ', ' Gross Sales ', ' Discounts ', '  Net Sales ',
       ' COGS ', ' Profit ']

# Replace non-numeric values (e.g., '$-', ' -   ') with NaN
df[columns_to_convert] = df[columns_to_convert].replace('[\$,]', '', regex=True).replace('', np.nan)

# Convert to float and set decimal places to 2 for each column
for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce').round(2)

df
```

Feature Engineering

Gross Profit Margin

```python
# df = df.assign(Profit_Margin=lambda x: (x[' Profit '] / x[' Gross Sales '] * 100).round(2))
df['Gross Profit Margin'] = (df[' Profit '] / df[' Gross Sales '] * 100).round(2)
df
```

Net Profit Margin

```python
# df = df.assign(Profit_Margin=lambda x: (x[' Profit '] / x[' Gross Sales '] * 100).round(2))
df['Net Profit Margin'] = (df[' Profit '] / df['  Net Sales '] * 100).round(2)
df
```

Segment 

```python
# Specify the certain text
segmentcode1 = 'PHC'
segmentcode2 = 'PTE'
segmentcode3 = 'TVG'
segmentcode4 = 'TMM'
segmentcode5 = 'SBS'

# Create a new column if the first three letters match the segment codes
df['Segment'] = np.where(df[' SKU no. '].str[:3] == segmentcode1, 'Channel Partners', '')
```

```python
# Update the existing column with the right text
df.loc[df[' SKU no. '].str[:3] == segmentcode2, 'Segment'] = 'Enterprise'
df.loc[df[' SKU no. '].str[:3] == segmentcode3, 'Segment'] = 'Government'
df.loc[df[' SKU no. '].str[:3] == segmentcode4, 'Segment'] = 'Midmarket'
df.loc[df[' SKU no. '].str[:3] == segmentcode5, 'Segment'] = 'Small Business'
df
```

```python
df['Segment'].value_counts()
```

Country

```python
# Specify the certain text
countrycode1 = 3
countrycode2 = 6
countrycode3 = 7
countrycode4 = 13
countrycode5 = 21

# Extract the number after the first three letters using regular expression
df['Extracted Number'] = df[' SKU no. '].str.extract(r'[^a-zA-Z](\d+)')

# Create a new column and input 'Yes' if the extracted number matches the target number
df['Country'] = np.where(df['Extracted Number'].astype(float) == countrycode1, 'Canada', '')
df
```

```python
# Update the existing column with the right text
df.loc[df['Extracted Number'].astype(float) == countrycode2, 'Country'] = 'France'
df.loc[df['Extracted Number'].astype(float) == countrycode3, 'Country'] = 'Germany'
df.loc[df['Extracted Number'].astype(float) == countrycode4, 'Country'] = 'Mexico'
df.loc[df['Extracted Number'].astype(float) == countrycode5, 'Country'] = 'United States of America'

df.head(3)
```

```python
df['Country'].value_counts()
```

Product Code

```python
# Specify the certain text
productcode1 = 'B'
productcode2 = 'C'
productcode3 = 'E'
productcode4 = 'O'
productcode5 = 'R'
productcode6 = 'V'

# Extract the last letter from the string
df['Last Letter'] = df[' SKU no. '].str[-1]

# Create a new column and input 'Yes' if the extracted number matches the target number
df['Product'] = np.where(df['Last Letter'] == productcode1, 'Basket Ballz', '')
df.head(3)
```

```python
# Update the existing column with the right text
df.loc[df['Last Letter'] == productcode2, 'Product'] = 'Cotton Ballz'
df.loc[df['Last Letter'] == productcode3, 'Product'] = 'Meat Ballz'
df.loc[df['Last Letter'] == productcode4, 'Product'] = 'Moth Ballz'
df.loc[df['Last Letter'] == productcode5, 'Product'] = 'Rapsa Ballz'
df.loc[df['Last Letter'] == productcode6, 'Product'] = 'Volley Ballz'
df.head(3)
```
